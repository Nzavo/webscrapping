Using BeautifulSoup

```python
url="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/netflix_data_webpage.html"
data=requests.get(url).text
```

```python
soup=BeautifulSoup(data, "html5lib")
```

```python
netflix_data=pd.DataFrame(columns=["Date", "Open", "High", "Low", "Close", "Adj Volume", "Volume"])

for row in soup.find("tbody").find_all("tr"):
    col=row.find_all("td")
    date=col[0].text
    open=col[1].text
    high=col[2].text
    low=col[3].text
    close=col[4].text
    adj_volume=col[5].text
    volume=col[6].text
    
#append the data of each row into the table
    netflix_data=netflix_data.append({"Date":date, "Open":open, "High":high, "Low":low, "Close":close, "Adj Volume":adj_volume, "Volume":volume},ignore_index=True)
    netflix_data
```
